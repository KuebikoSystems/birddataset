```python
import matplotlib.pyplot as plt
import librosa 
import librosa.display
import IPython.display as ipd
```

```python
# Importing processing libraries
# Bird Number 1 :Common Rock Thrush
cm_path = "/content/birddataset/Species/Common Rock Thursh/XC102155 - Common Rock Thrush - Monticola saxatilis.mp3"
ct_path = "/content/birddataset/Species/Crested Tit/XC246413 - European Crested Tit - Lophophanes cristatus.mp3"
rr_path ="/content/birddataset/Species/Red-rumped Swallow/XC397895 - Red-rumped Swallow - Cecropis daurica.mp3"
```

```python
x1,sr1 = librosa.load(cm_path, sr= 44100)
x2,sr2 = librosa.load(ct_path, sr= 44100)
x3,sr3 = librosa.load(rr_path, sr= 44100)
# print (type(x),type(sr))
ipd.Audio(cm_path)
ipd.Audio(ct_path)
ipd.Audio(rr_path)

```

```python
# display waveform
plt.figure(figsize =(14,6))
librosa.display.waveplot(x1,sr=sr1)
plt.figure(figsize =(14,6))
librosa.display.waveplot(x2,sr=sr2)
plt.figure(figsize =(14,6))
librosa.display.waveplot(x3,sr=sr3)

```

```python
# Display spectrogram for Common Rock Thrush
X1= librosa.stft(x1)
X1db = librosa.amplitude_to_db(abs(X1))
plt.figure(figsize=(14,6))
librosa.display.specshow(X1db, sr=sr1, x_axis='time', y_axis='hz')  
#librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')  #If to pring log of frequencies
plt.colorbar()
plt.title("Common Rock Thrush")

# Display spectrogram for European Crested Tit
X2= librosa.stft(x2)
X2db = librosa.amplitude_to_db(abs(X2))
plt.figure(figsize=(14,6))
librosa.display.specshow(X2db, sr=sr2, x_axis='time', y_axis='hz')  
#librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')  #If to pring log of frequencies
plt.colorbar()
plt.title("European Crested Tit")

# Display spectrogram for Red Rumped Swallow
X3= librosa.stft(x2)
X3db = librosa.amplitude_to_db(abs(X3))
plt.figure(figsize=(14,6))
librosa.display.specshow(X3db, sr=sr3, x_axis='time', y_axis='hz')  
#librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')  #If to pring log of frequencies
plt.colorbar()
plt.title("Red Rumped Swallow")
```
